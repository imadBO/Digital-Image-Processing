```python
import numpy as np
import cv2
import matplotlib.pyplot as plt 
import math

def ReadImage (argument) :
    img = cv2.imread(argument,cv2.IMREAD_GRAYSCALE)
    return img 

def Expansion (argument) :
    img = ReadImage(argument)
    expansion = ((img.astype(int) - np.amin(img)))  * 255 / (np.amax(img) - np.amin(img))
    #cv2.imwrite('Expansion.png',expansion)
    return expansion

def Egalisation (argument) :
    img = ReadImage(argument)
    egalisation = cv2.equalizeHist(img)
    #cv2.imwrite('galisation.png',egalisation)
    return egalisation

Images = ['Fig0208(a).tif','Fig0210(a).tif','Fig0228(a).tif']


for i in Images :
    fig = plt.figure()
    fig.set_size_inches(19 , 10)
    ax1 = fig.add_subplot(1,3,1)
    ax1.set_xlabel("originale")
    ax1.imshow(ReadImage(i), cmap = 'gray' ,vmin= 0,vmax = 255)
    ax1 = fig.add_subplot(1,3,2)
    ax1.set_xlabel("Expansion")
    ax1.imshow(Expansion(i) , cmap = 'gray', vmin= 0,vmax = 255)
    str = "Expansion Image "+ i 
    plt.imsave(str+".png", Expansion(i), cmap='gray',vmin=0,vmax=255)
    ax1 = fig.add_subplot(1,3,3)
    ax1.set_xlabel("Egalisation")
    ax1.imshow(Egalisation(i) , cmap = 'gray' ,vmin= 0,vmax = 255)
    str2 = "equalized Image " + i
    plt.imsave(str2+".png", Egalisation(i), cmap='gray',vmin=0,vmax=255)

```
